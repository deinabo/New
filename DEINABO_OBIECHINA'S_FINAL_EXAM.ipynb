```
from sklearn.decomposition import PCA
pca = PCA(n_components=6)

pca.fit(hr_numeric.values) 

hr_numeric_pca = pca.fit_transform(hr_numeric[['employee_id', 'no_of_trainings', 'age', 'previous_year_rating', 'length_of_service','avg_training_score']].values) 

#let's check the shape of df_new_pca array
print("shape of hr_numeric_pca", hr_numeric_pca.shape)
```

```
const = ["E1", "E2", "E3", "E4", "E5", "E6"]
len(const)
```

```
pca_exp_var = pd.DataFrame(pca.explained_variance_, index = const, columns = ["variance"])

pca_pctexp_var = pd.DataFrame(pca.explained_variance_ratio_ * 100, index = const, columns = ["% variance"])

pca_exp_var.merge(pca_pctexp_var, left_index = True, right_index = True)
```

```
pca_hr = pd.DataFrame(pca.components_, columns = const, index = hr_numeric.columns.to_list())
pca_hr
```
